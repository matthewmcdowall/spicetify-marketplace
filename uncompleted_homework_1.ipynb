Here's a context-free grammar for a fragment of regular English:

Phrase rules:
* **NP → NN**  e.g. "Frank"
* **NP → JJ NP**  e.g. "grumpy cat"
* **NP → DET NP**  e.g. "the cat"
* **PP → PRP NP** e.g. "from the cat"

* **VP → V NP** e.g. "feeds the cat"
* **VP → V PP** e.g. "runs from the cat"
* **S → NP VP** e.g. "Frank runs from the cat"

### Visually inspecting some vectors (10 points)

#### Preliminary stuff (don't change this)

I wrote you a little function that takes in a (preprocessed) text, uses one of your fitted vectorizers, and produces a visualization of the vector produced for that text, by that vectorizer.

````python
from typing import Union
def display_vectorization(text:str,
                          vectorizer:Union[CountVectorizer, TfidfVectorizer]):
  rval = []
  v = vectorizer.transform([text])
  vocab = vectorizer.get_feature_names_out()
  for row, column in zip(*v.nonzero()):
    rval.append({'word':vocab[column], 'value':v[row,column]})

  return pd.DataFrame(rval)
````

Now I run it on a little toy movie review, for both your CountVectorizer and your TfidfVectorizer. Take a look at the difference

````python
text = "It is a really really terrible movie!"
print('Review text:', text)

preprocessed_text = preprocess(text, stem=True)
print ('\nPreprocessed review text:', preprocessed_text)

print('\nCount vector:')
display(display_vectorization(preprocessed_text, count_vectorizer))

print('\nTF-IDF vector:')
display(display_vectorization(preprocessed_text, tfidf_vectorizer))
````
